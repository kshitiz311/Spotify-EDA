#### Setting the index to release_date

```python
df.set_index("release_date", inplace=True)
df.index=pd.to_datetime(df.index)
df.head()
```

### Checking an artist at a specific position (or row)

```python
df[["artists"]].iloc[18]
```

#### Converting song duration from milliseconds to seconds 

```python
df['duration']=df['duration_ms'].apply(lambda x : round(x/1000))
df.drop('duration_ms', inplace=True, axis=1)
df.duration.head()
```

### Visualizing through a correlation map

```python
corr_df=df.drop(['key','mode','explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14,8))
heatmap=sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap="BrBG", linewidths=2, linecolor="Black"
)
heatmap.set_title('Correlation Heatmap Between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

#### Sampling the dataset
to sample like 0.4% of the original dataset

```python
sample_df=df.sample(int(0.004*len(df)))
print(len(sample_df))
```

### Regression plot between Loudness and Energy

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='loudness', x='energy', scatter_kws={"color": "violet"}, line_kws={"color": "black"}).set(title='Loudness Vs Energy Correlation')
```

### Regression plot between Popularity and Acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='popularity', x='acousticness', scatter_kws={"color": "c"}, line_kws={"color": "black"}).set(title='Popularity Vs Acousticness Correlation')
```

### Regression plot between Speechiness and Acousticness

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='speechiness', x='acousticness', scatter_kws={"color": "goldenrod"}, line_kws={"color": "black"}).set(title='Speechiness Vs Acousticness Correlation')
```

##### Extracting years from "dates" column and creating a new "years" column

```python
df['dates']=df.index.get_level_values('release_date')
df.dates=pd.to_datetime(df.dates)
years=df.dates.dt.year
```

### Distribution plot to visualize the total number of songs in each year since 1922 in the Spotify database

```python
sns.displot(years, discrete=True , aspect=3, height=6 , kind='hist').set(title='Number Of Song Per Years')
```

### Bar plot to visualize the duration of songs over the years

```python
total_dr=df.duration
fig_dims=(20,8)
fig, ax =plt.subplots(figsize= fig_dims)
fig=sns.barplot(x=years,y=total_dr, ax=ax,errwidth=False).set(title='Year Vs Duration')
plt.xticks(rotation=90)
```

### Reading Spotify Features dataset

```python
df_genre = pd.read_csv('D:/DATA ANALYST PROJECTS/Spotify Exploratory Project/SpotifyFeatures.csv')
df.head()
```

### Bar plot to visualize the correlation between the duration of songs and their different genres

```python
plt.title("Duration of the Songs in Different Genres")
sns.color_palette("rocket", as_cmap= True)
sns.barplot(y='genre', x='duration_ms', data=df_genre)
plt.xlabel ("Duration in milli seconds")
plt.ylabel("Genres")
```

### Top5 genres by popularity

```python
sns.set_style(style = "darkgrid")
plt.figure(figsize=(10,5))
famous = df_genre.sort_values("popularity", ascending = False).head(10)
sns.barplot(y='genre', x='popularity', data = famous, errwidth = False).set(title= "Top 5 Genres by Popularity")
```
